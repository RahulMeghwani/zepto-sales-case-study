```python
## 1. Top 5 products by quantity sold

import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('zepto_sales_dataset.csv')
top_products_qty = df.groupby('Product Name')['Orders'].sum().sort_values(ascending=False).head(5)
print("Top 5 Products by Quantity Sold:")
print(top_products_qty)


top_products_qty.plot(kind='bar', color='skyblue', figsize=(8,5))
plt.title("Total Quantity Sold per Product")
plt.xlabel("Product Name")
plt.ylabel("Quantity Sold")
plt.xticks(rotation=45) 
plt.tight_layout() 
```

```python
## 2. Top Products by Revenue in Each City

df = pd.read_csv('zepto_sales_dataset.csv')
grouped = df.groupby(['City', 'Product Name'])['Total Revenue'].sum().reset_index()
sorted_df = grouped.sort_values(['City', 'Total Revenue'], ascending=[True, False])
top_per_city = sorted_df.groupby('City').head(3)
print("Top 3 Products by revenue in each city:")
print(top_per_city)

plt.figure(figsize=(10, 6))
sns.barplot(
    data=top_per_city,
    x='City',
    y='Total Revenue',
    hue='Product Name'  # Separate bars for each product
)
plt.xticks(rotation=45)
```

```python
## 3.	Products with High Revenue but Low Quantity Sold

df = pd.read_csv("zepto_sales_dataset.csv")
grouped1 = df.groupby('Product Name')[['Total Revenue','Orders']].sum().reset_index()
grouped1['Revenue per Unit'] = grouped1['Total Revenue'] / grouped1['Orders']
sorted_df = grouped1.sort_values(by='Revenue per Unit', ascending=False)
sorted_df1 = grouped1.sort_values(by='Orders', ascending=False)
print(sorted_df)
print("")
print(sorted_df1)

plt.figure(figsize=(10,6))
sns.scatterplot(data=grouped1, x='Orders', y='Revenue per Unit')

plt.title('Revenue per Unit vs Orders')
plt.xlabel('Orders')
plt.ylabel('Revenue per Unit')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## 5.	City-Wise Average Order Value

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("zepto_sales_dataset.csv")
total_revenue_per_city = df.groupby('City')['Total Revenue'].sum().reset_index()
rearranging_revenue = total_revenue_per_city.sort_values(by="Total Revenue",ascending=False)
print(rearranging_revenue)
print("")
total_orders_per_city = df.groupby('City')['Orders'].sum().reset_index()
rearranging_orders = total_orders_per_city.sort_values(by="Orders",ascending=False)
print(rearranging_orders)
print("")
merged = pd.merge(total_revenue_per_city, total_orders_per_city, on='City')
merged['AOV'] = merged['Total Revenue'] / merged['Orders']
merged = merged.sort_values(by='AOV', ascending=False)
print(merged)
print("")

sns.barplot(x='City', y='AOV', data=merged, palette = "pastel")
plt.xticks(rotation=45)
plt.title("City-wise Average Order Value")
plt.tight_layout()
plt.show()
```
